```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dtype={'user_id': int}
```

```python
# Load the dataset
data = pd.read_csv('/Users/harshkansara/Downloads/globalterrorismdb_0718dist.csv', encoding='latin1')
```

```python
# Display basic information about the dataset
data.info()
```

```python
# Display first few rows of the dataset
data.head()
```

```python
# Cleaning the data
# Removing unnecessary columns
data = data[['iyear','imonth','iday','country_txt','region_txt','city','latitude','longitude','success','attacktype1_txt','targtype1_txt','weaptype1_txt','nkill','nwound']]
```

```python
# Renaming columns for better understanding
data.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'region_txt':'Region', 'city':'City', 'latitude':'Latitude', 'longitude':'Longitude', 'success':'Success', 'attacktype1_txt':'Attack_Type', 'targtype1_txt':'Target_Type', 'weaptype1_txt':'Weapon_Type', 'nkill':'Fatalities', 'nwound':'Injuries'}, inplace=True)
```

```python
# Handling missing values
data.dropna(inplace=True)
```

```python
# Plotting the number of terrorist attacks each year
plt.figure(figsize=(15, 6))
sns.countplot(data['Year'])
plt.title('Number of Terrorist Attacks Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

### Number of Terrorist Attacks Each Year: This graph shows the trend of terrorist attacks over the years, helping to identify any patterns or significant changes over time.

```python
# Plotting the distribution of terrorist attacks by region
plt.figure(figsize=(12, 8))
sns.countplot(y=data['Region'], order=data['Region'].value_counts().index)
plt.title('Distribution of Terrorist Attacks by Region')
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.show()
```

### Distribution of Terrorist Attacks by Region: This graph provides insight into which regions are most affected by terrorism, crucial for understanding regional security concerns.

```python
# Plotting the number of terrorist attacks by country
plt.figure(figsize=(15, 8))
sns.countplot(y=data['Country'], order=data['Country'].value_counts().iloc[:20].index)
plt.title('Top 20 Countries by Number of Terrorist Attacks')
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()
```

### Top 20 Countries by Number of Terrorist Attacks: This graph highlights the countries with the highest number of terrorist attacks, indicating areas requiring heightened security measures.

```python
# Plotting the distribution of attack types
plt.figure(figsize=(12, 8))
sns.countplot(y=data['Attack_Type'], order=data['Attack_Type'].value_counts().index)
plt.title('Distribution of Attack Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Attack Type')
plt.show()
```

### Distribution of Attack Types: Understanding the distribution of attack types helps in strategizing defense mechanisms tailored to different attack scenarios.

```python
# Plotting the distribution of target types
plt.figure(figsize=(12, 8))
sns.countplot(y=data['Target_Type'], order=data['Target_Type'].value_counts().index)
plt.title('Distribution of Target Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Type')
plt.show()
```

### Distribution of Target Types: Knowing the types of targets terrorists commonly choose helps in protecting vulnerable groups and locations.

```python
# Plotting the distribution of weapon types
plt.figure(figsize=(12, 8))
sns.countplot(y=data['Weapon_Type'], order=data['Weapon_Type'].value_counts().index)
plt.title('Distribution of Weapon Types')
plt.xlabel('Number of Attacks')
plt.ylabel('Weapon Type')
plt.show()
```

### Distribution of Weapon Types: This graph illustrates the weapons most commonly used by terrorists, guiding efforts to intercept weapon supply chains and enhance border security.

```python
# Plotting the correlation between fatalities and injuries
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Fatalities', y='Injuries', data=data)
plt.title('Correlation between Fatalities and Injuries')
plt.xlabel('Fatalities')
plt.ylabel('Injuries')
plt.show()
```

### Correlation between Fatalities and Injuries: Analyzing the correlation between fatalities and injuries provides insight into the severity of terrorist attacks and helps in allocating resources for emergency response and medical care. 

```python
# Finding the hot zone of terrorism
hot_zone = data['Country'].value_counts().idxmax()
print(f'The hot zone of terrorism is {hot_zone}')
```
